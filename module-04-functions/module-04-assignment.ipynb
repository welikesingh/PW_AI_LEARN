```python
#Q1: write a python function that takes a list of numbers as input and return the sum of all even numbers in the list  

def sum_even(p):
    s=0
    for e in p:
        if e%2==0:
            s+=e
    return s
list1=[2,3,4]        
sum_even(list1)
```

```python
#Q3 impliment a python function that takes a list of integers and return a new list containing the square of each number
n=[1,2,3,4]
def sqr(p):
    n_square=[]
    for i in p:
        n_square.append(i*i)
    #print(n_square)
    return n_square

print(sqr(n))
```

```python
#Q4 write a python function that check if a given number is prime or not from 1 to 200
def my_prime_check(n):
    for i in range(n): 
        if i not in [0,1,n] and n%i == 0:
            return False
    return True

my_prime_check(12)

```

```python
#5 create an iterator class in python that generate fibonacci series upto a specified number of terms
def fib(n):
    a = 0 
    b = 1
    for i in range(n): #if n=5, 0, 1, 2, 3, 4
        yield a
        a, b = b, a+b
f = fib(100)
f
```

```python
#q6 write a generator function in python that yield power of 2 upto a given exponenet
x=int(input("Please enter exponent:"))
#method 1
def power_of_two(p):
    for i in range(1,p):
        yield 2**i
f = power_of_two(x) 
for i in range(1,x):     
   print(next(f))


#method 2
x_list=[i for i in range(1,x)]
sqr_num= map(lambda a: 2**a, x_list)
print(list(sqr_num))
```

```python

#Q7 write a generator function that read a file line by line and yield each line as string
# f = open("sample.txt", "r")
# print(f.read())

file_name="sample.txt"
def print_lines(file_name):
    # f = open(file_name, "r")
    # print(f.read())
    with open(file_name,'r') as file:
        for line in file:
            yield line.strip()  # Using strip() to remove trailing newline characters
for i in print_lines(file_name):
    print(i)
```

```python
#Q8 Use a lambda function in python to sort a list of tuples based on the second element of tuple

tuples_list=((1,4,5), (8,20,3), (8,9,2))
sorted_list = sorted(tuples_list, key=lambda x: x[1])
print(sorted_list)
```

```python
#q10 create a python program that uses filter to remove vowels from given string
# letter representing a vowel sound, such as a, e, i, o, u.
given_list= [i for i in "balwant"]
print(given_list)

#method 1
def is_vowel(p):
    if (p.lower() in ['a','e','i','o','u']):
        return False
    else:
        return True

print(list(filter(is_vowel,given_list)))   


#metho 2
mod_list=filter(lambda a: a.lower() not in ['a','e','i','o','u'] ,  given_list)
print(list(mod_list))
```
